```python
## Ques-3. Write code that will print out the anagrams (words that use the same letters) from a paragraph of text.
```

```python
paragraph = "Hi, my name is mane si"

word_list = paragraph.split(' ')
print(word_list)

print("The words that are anagram in above paragraph are listed below:")
for i in range(len(word_list)):
    
    for j in range(i, len(word_list)-1):
        first_word = list(word_list[i])
        actual_word = ''.join(first_word)
        first_word.sort()
        second_word = list(word_list[j+1])
        second_word.sort()
        
        if first_word == second_word:
            print('>',actual_word)
```

```python
## Ques-7. Create a list of tuples of first name, last name, and age for your friends and colleagues. If you don't know the
## age, put in None. Calculate the average age, skipping over any None values. Print out each name, followed by old or young if 
## they are above or below the average age.
```

```python
from functools import reduce

people_data = [('Deepika', 'Gurung', 28), ('Aakriti', 'Bista', 22), ('Rashika', 'Shrestha', 34), ('Sanju', 'Pradhan', 42)]

ages = [x[2] for x in people_data]
sum_of_ages = 0
age_count = 0
for age in ages:
    if age is not None:
        age_count += 1
        sum_of_ages += age

avg_age = sum_of_ages / age_count

for first_name, last_name, age in people_data:
    print(f'{first_name} {last_name}', end=' ')
    if age is not None:
        if age < avg_age:
            print('young')
        else:
            print('old')
    else:
        print('Unknown')
```

```python
## Ques-8. Write a function, is_prime, that takes an integer and returns True if the number is prime and False if the number 
## is not prime.
```

```python
def test_prime(n):
    if (n==1):
        return False
    elif (n==2):
        return True;
    else:
        for x in range(2,n):
            if(n % x==0):
                return False
        return True             
print(test_prime(17))
```

```python
## Ques-9. Write a binary search function. It should take a sorted sequence and the item it is looking for. It should return
## the index of the item if found. It should return -1 if the item is not found.
```

```python
sequence = [i for i in range(1, 30)]
sequence.sort()
search_item = 20

def search(seq, item):
    try:
        a = seq.index(item)
    except ValueError:
        return -1
    return a

print(search(sequence, search_item))
```

```python
## Ques-10. Write a function that takes camel-cased strings (i.e. ThisIsCamelCased), and converts them to snake case (i.e.
## this_is_camel_cased). Modify the function by adding an argument, separator, so it will also convert to the kebab case
## (i.e.this-is-camel-case) as well.
```

```python
import re

# convert the camel cased string to snake case
def camel_cased_to_snake(strg):
    strg = re.sub(r'(?<!^)(?=[A-Z])', '_', strg).lower()
    return strg

# convert the camel cased string to kebab case
def camel_cased_to_kebab(strg):
    strg = re.sub(r'(?<!^)(?=[A-Z])', '-', strg).lower()
    return strg

print("Snake Case = {} \nKebab Case = {}"
        .format(camel_cased_to_snake("ThisIsCamelCased"), camel_cased_to_kebab("ThisIsCamelCased")))
```

```python
## Ques-12. Create a function, is_palindrome, to determine if a supplied word is the same if the letters are reversed.
```

```python
samp_word = 'Madam'

def is_palindrome(word):
    reversed_word = word[::1]
    if word == reversed_word:
        print(f'The word {word} is palindrome.')
    else:
        print(f'The word {word} is not palindrome.')

is_palindrome(samp_word)
```

```python
## Ques-13. Write a function to write a comma-separated value (CSV) file. It should accept a filename and a list of tuples as 
## parameters. The tuples should have a name, address, and age. The file should create a header row followed by a row for each 
## tuple. If the following list of tuples was passed in:
## [('George', '4312 Abbey Road', 22), ('John', '54 Love Ave', 21)] it should write the following in the file:
## name,address,age
## George,4312 Abbey Road,22
## John,54 Love Ave,21
```

```python
import csv
details_list = [('George', '4312 Abbey Road', 22), ('John', '54 Love Ave', 21)]

def write_csv(sample_list):
    with open('info_file.csv', 'w', newline='') as personal_details:
        detail_writer = csv.writer(personal_details)
        header_row = ['name', 'address', 'age']
        detail_writer.writerow(header_row)
        for detail in sample_list:
            detail_writer.writerow(list(detail))
    personal_details.close()

write_csv(details_list)
```

```python
## Ques-14. Write a function that reads a CSV file. It should return a list of dictionaries, using the first row as key names, 
## and each subsequent row as values for those keys. For the data in the previous example it would return:
## [{'name': 'George', 'address': '4312 Abbey Road', 'age': 22}, {'name':'John', 'address': '54 Love Ave', 'age': 21}]
```

```python
import csv

def read_data(file_name):
    result_data_list = []
    with open(file_name, newline='') as file:
        file_reader = csv.reader(file, delimiter=',')
        is_first_row = True
        for data in file_reader:
            temp_dict = {}
            if is_first_row:
                keys = data
                is_first_row = False
                continue
            for i in range(len(keys)):
                temp_dict[keys[i]] = data[i]
            result_data_list.append(temp_dict)
    file.close()
    return result_data_list


print(read_data('info_file.csv'))
```

```python
## Ques-15. Imagine you are designing a banking application. What would a customer look like? What attributes would she have? 
## What methods would she have?
```

```python
class CustomerInfo:

    customerList = [{"Name": "Roshan Pradhan", "Account No.": "123014789612", "Gender": "Male", "Balance": 800000}]
    minBalance = 20000

    def __init__(self, accName, accNumber):
        self.accName = accName
        self.accNumber = accNumber
        
    def getInfo(self):
        customer = next((item for item in self.customerList if item["Account No."] == self.accNumber and 
                         item["Name"] == self.accName), None)
        if (customer is not None):
            print(f"Details of the Customer: \nA/C: {customer.get('Account No.')} \nOwner: {customer.get('Name')} \nGender: {customer.get('Gender')} \nBalance: {customer.get('Balance')}")
        else:
            print(f"No customer {self.accName} found")

    @classmethod
    def balance(classObject):
        return classObject.minBalance

customer = CustomerInfo("Roshan Pradhan", "123014789612")
customer.getInfo()

print("Minimum balance:", CustomerInfo.balance())
```

```python
## Ques-16. Imagine you are creating a Super Mario game. You need to define a class to represent Mario. What would it look 
## like? If you aren't familiar with SuperMario, use your own favorite video or board game to model a player.
```

```python
class SuperMario:

    name = "Mario"
    color = "Red"

    def __init__(self):
        self.level = 1
        self.move = self.Movements()

    def display(self):
        maxlevel = 5
        if self.level < maxlevel:
            print("Welcome to Level",self.level, "of Super Mario")
        else:
            print("Sorry, we have only upto level", maxlevel)

        self.move.displayMovements()
        self.move.levelComplete()

    @classmethod
    def characterInfo(cls):
        print("Your character name is {}. \nYour character color is {}."
                .format(cls.name,cls.color))

    class Movements:
        def __init__(self):
            self.move_right,self.move_left = 8, 9
            self.go_up, self.go_down = 4, 7

        def displayMovements(self):
            print("Move Right = {} \nMove Left = {} \nJump = {} \nDown = {}"
                        .format(self.move_right,self.move_left,self.go_up,self.go_down))

        def levelComplete(self):
            if self.move_right + self.move_left + self.go_down + self.go_up >= 20:
                print("Level Completed")
            elif self.move_right + self.move_left + self.go_down + self.go_up < 20:
                print("Advance to complete this level")


mario = SuperMario()
SuperMario.characterInfo()
mario.display()
```

```python
## Ques-17. Write a program that serves as a basic calculator. It asks for two numbers, then it asks for an operator. 
## Gracefully deal with input that doesn't cleanly convert to numbers. Deal with division by zero errors.
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y


print("Select Operation:")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")

while True:
    # Take input from the user
    choice = input("Enter choice(1/2/3/4): ")

    # Check if choice is one of the four options
    if choice in ('1', '2', '3', '4'):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print(num1, "+", num2, "=", add(num1, num2))

        elif choice == '2':
            print(num1, "-", num2, "=", subtract(num1, num2))

        elif choice == '3':
            print(num1, "*", num2, "=", multiply(num1, num2))

        elif choice == '4':
            print(num1, "/", num2, "=", divide(num1, num2))
        break
    else:
        print("Invalid Input")
```

```python
## Ques-19. Write a Python class to find validity of a string of parentheses, '(',')', '{', '}', '[' and ']. These brackets 
## must be close in the correct order, for example "()" and "()[]{}" are valid but "[)", "({[)]" and "{{{" are invalid.
```

```python
class Parentheses:
    def is_valid(self, value):
        stack, brackets = [], {"(": ")", "{": "}", "[": "]"}
        for parentheses in value:
            if parentheses in brackets:
                stack.append(parentheses)
            elif len(stack) == 0 or brackets[stack.pop()] != parentheses:
                return False
        return len(stack) == 0

print(Parentheses().is_valid("()"))
print(Parentheses().is_valid("(){}[]"))
print(Parentheses().is_valid("[)({[)]"))
print(Parentheses().is_valid("{{{"))
```

```python
## Ques-20. Write a Python class to find the three elements that sum to zero from a list of n real numbers.
## Input array : [-25, -10, -7, -3, 2, 4, 8, 10]
## Output : [[-10, 2, 8], [-7, -3, 10]]
```

```python
class Sum:

    def sum_to_zero(self, number):
        number = sorted(number)
        result = []
        i = 0
        while i < len(number) - 2:
            j, k = i + 1, len(number) - 1
            while j < k:
                if number[i] + number[j] + number[k] < 0:
                    j += 1
                elif number[i] + number[j] + number[k] > 0:
                    k -= 1
                else:
                    result.append([number[i], number[j], number[k]])
                    j ,k = j + 1, k - 1
                    while j < k and number[j] == number[j - 1]:
                        j += 1
                    while j < k and number[k] == number[k + 1]:
                        k -= 1
            i += 1
            while i < len(number) - 2 and number[i] == number[i - 1]:
                i += 1
        return result

print("The output is:")
print(Sum().sum_to_zero([-25, -10, -7, -3, 2, 4, 8, 10]))
```
